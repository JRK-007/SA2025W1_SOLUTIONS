### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
import pandas as pd

# Load dataset
df = pd.read_csv('Cars.csv')

# 3. Most common origin for cars with horsepower > 100 and weight < 3000
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()[0]
print("Most common origin (hp > 100 & weight < 3000):", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
import pandas as pd

# Load dataset
df = pd.read_csv('Cars.csv')

# Strip any whitespace from column names just in case
df.columns = df.columns.str.strip()

# Calculate the median mpg
median_mpg = df['mpg'].median()

# Filter cars with mpg above the median
high_mpg_cars = df[df['mpg'] > median_mpg]

# Calculate horsepower-to-weight ratio
high_mpg_cars['hp_per_weight'] = high_mpg_cars['horsepower'] / high_mpg_cars['weight']

# Find the max ratio value
max_ratio = high_mpg_cars['hp_per_weight'].max()

# Get the car(s) with the best ratio
best_cars = high_mpg_cars[high_mpg_cars['hp_per_weight'] == max_ratio]

print("Car(s) with best horsepower-to-weight ratio among cars with mpg above median:")
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Cars.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Verify column names (optional, can comment out after confirming)
print(df.columns.tolist())

# Group by year and origin to get average mpg
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Load dataset
df = pd.read_csv('Cars.csv')
df.columns = df.columns.str.strip()

# Step 1: Identify car names that appear in more than one model_year
name_counts = df.groupby('name')['model_year'].nunique()
multi_year_names = name_counts[name_counts > 1].index

# Filter dataframe to only those cars
df_multi = df[df['name'].isin(multi_year_names)]

# Step 2: For each such name, compute std dev of mpg, number of appearances, and average mpg
stats = df_multi.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    appearances=('model_year', 'nunique'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Step 3: Filter those with std(mpg) < 1.0
consistent_cars = stats[stats['mpg_std'] < 1.0]

# Step 4: Find the car(s) with the lowest variation in mpg
min_std = consistent_cars['mpg_std'].min()
lowest_variation_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

# Bonus: Sort by number of appearances (desc), then avg mpg (desc)
result = lowest_variation_cars.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

print(result[['name', 'appearances', 'avg_mpg']])
```
